```python

import flet as ft
import time


def main(pagina):
    texto = ft.Text('Bem-vindo ao NewChat', size= 18)
    texto_chat_ok = ft.Text('Chat on-line')   
    
    chat = ft.Column()
    
    nome_usuario = ft.TextField(label="Name")
    
    def enviar_msg_tunel(mensagem):
        tipo = mensagem["tipo"]
        if tipo == "mensagem":
            texto_mensagem = mensagem["texto"]
            usuario_mensagem = mensagem["usuario"]
            #add msg 
            chat.controls.append(ft.Text(f"{usuario_mensagem}:{texto_mensagem}"))
        else:
            usuario_mensagem = mensagem["usuario"]
            chat.controls.append(ft.Text(f"{usuario_mensagem} entrou no chat", size= 15, italic=True, color=ft.colors.BLUE))
        pagina.update()

    #tuneo de envio de mensagem
    pagina.pubsub.subscribe(enviar_msg_tunel)

    def enviar_msg(evento):
        pagina.pubsub.send_all({"texto": campo_mensagem.value, "usuario": nome_usuario.value, "tipo": "mensagem"})
        #limpar campo msg
        campo_mensagem.value = ""
        pagina.update()


    campo_mensagem = ft.TextField(label="Digite uma mensagem", on_submit= enviar_msg)
    botao_enviar_msg = ft.ElevatedButton("Enviar", on_click= enviar_msg)


    def entrar_popup(evento):
        pagina.pubsub.send_all({"usuario" : nome_usuario.value, "tipo": "entrada"})
        #adicionar o chat
        pagina.add(chat)
        #fechar o popup
        popup.open=False
        #remover o botão iniciar da pagina
        pagina.remove(botao_iniciar)
        #remover o texto de início 
        pagina.remove(texto)
        #Adiciona novo texto de início
        pagina.add(texto_chat_ok)
        #criar o campo de mensagem do usuário
        #pagina.add(campo_mensagem)
        #criar botão de enviar msg d usuário
        pagina.add(ft.Row([campo_mensagem, botao_enviar_msg]))
        #atualizar pagina
        pagina.update()



    popup = ft.AlertDialog(
        open=False,
        modal=True,
        title=ft.Text("Digite seu nickname"),
        content=nome_usuario,
        actions=[ft.ElevatedButton("Entrar", on_click=entrar_popup)],
    )


    def entrar_chat(evento):
        pagina.dialog = popup
        popup.open = True
        pagina.update()

    botao_iniciar = ft.ElevatedButton('Iniciar', on_click=entrar_chat)
    
    pagina.add(texto)
    pagina.add(botao_iniciar)
    
    
ft.app(target=main)
#ft.app(target=main, view=ft.WEB_BROWSER, port=8000)



#cmd
#ipconfig
#endereço IPV4
 #Endereço IPv4. . . . . . . .  . . . . . . . : 172.16.3.95
 
 #172.16.3.95:8000
```
